```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\conno_leva1li\Downloads\data_jobs.csv')

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skill: ast.literal_eval(skill) if pd.notna(skill) else skill )
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country']=='United States')]
df_DA_US['job_posted_month'] = df_DA_US['job_posted_date'].dt.month
df_DA_US_skill = df_DA_US.explode('job_skills')

# df_DA_US_skill['job_skills'].value_counts().reset_index(name='job_count').sort_index().head(7)
# top_skills = df_DA_US_skill['job_skills'].value_counts().reset_index(name='job_count').sort_index().head(7).values
# top_skills
```

```python
df_pivot = df_DA_US_skill.pivot_table(index='job_posted_month', columns='job_skills', aggfunc='size', fill_value=0)
df_pivot.loc['Total'] = df_pivot.sum()
df_pivot = df_pivot[df_pivot.loc['Total'].sort_values(ascending=False).index]
df_pivot = df_pivot.drop('Total')
df_pivot
```

```python
DA_totals = df_DA_US.groupby('job_posted_month').size()
DA_totals
```

```python
df_DA_US_percent = df_pivot.div(DA_totals/100, axis=0)
```

```python
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month_abv'] = df_DA_US_percent['job_posted_month'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month_abv')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month')
df_DA_US_percent
```

```python
df_plot = df_DA_US_percent.iloc[:, 0:5]
df_plot
```

```python
sns.lineplot(data=df_plot, dashes=False)
sns.set_theme(style='ticks')
plt.title('Trending Top Skills for Data Analyst in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
sns.despine()
for i in range(5):
    if i == 3:
        plt.text(10.6, df_plot.iloc[-2, i], df_plot.columns[i])
    else:
        plt.text(11.1, df_plot.iloc[-1, i], df_plot.columns[i])
plt.text(10.3, df_plot.iloc[-2, 3], '--')
ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{x:.0f}%'))
```
