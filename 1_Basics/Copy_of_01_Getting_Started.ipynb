Just be aware that starting a line with any of the following symbols has magic properties.

| Symbol | Name           | Example            | Usage Explanation                                                                 |
|--------|----------------|--------------------|-----------------------------------------------------------------------------------|
| `!`    | Shell Command  | `!ls`              | Executes shell commands directly from the notebook cell, such as listing files.   |

| `%%`   | Cell Magic     | `%%writefile test.txt` | Applies the magic command to the entire cell, such as writing the cell's content to a file. |

```python
import pandas as pd

# Try reading the CSV with the 'python' engine to handle potential parsing issues
df = pd.read_csv(r'C:\Users\conno_leva1li\Downloads\data_jobs.csv')

# Display the first few rows of the DataFrame
display(df.head())
df.info()
```

```python
# Select rows where 'job_title' contains 'Data' and 'salary_year_avg' is not NaN, and then select the 'salary_year_avg' column
selected_salaries = df.loc[(df.job_title.str.contains('Data')) & (df.salary_year_avg.notna()), 'salary_year_avg']

# Display the selected data
display(selected_salaries)
```

```python
df.sort_values(by='salary_year_avg', ascending=False).head()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df.info()
```

```python
df.describe()
```

```python
df.groupby(['job_title_short'])['salary_year_avg'].median()
```

```python
import matplotlib.pyplot as plt
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
monthly_counts = df.job_posted_month.value_counts()
```

```python
monthly_counts = monthly_counts.sort_index()
monthly_counts
```

```python
from datetime import date
plt.plot(monthly_counts.index, monthly_counts)
```

```python
job_count = df.job_title_short.value_counts()
```

```python
job_count = job_count.sort_values(ascending=False)
plt.bar(job_count.index, job_count)
plt.title('Job Title Counts')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation = 45, ha='right')
```

```python
job_count.plot(kind='bar')

plt.title('Job Title Counts')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation = 45, ha='right')
plt.xlabel('')
```

```python
df.groupby('job_title_short')['salary_year_avg'].median().sort_values().plot(kind='barh')
plt.title('Median Salary by Job Title')
plt.xlabel('Salary in USD')
plt.ylabel('')
```
